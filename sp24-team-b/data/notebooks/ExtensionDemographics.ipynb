```python
import matplotlib.pyplot as plt
```

```python
# Dorchester: #1 Neighborhood with Most Building Violations (Normalized)
categories = ['White', 'Black/ African American', 'Hispanic/ Latino',
              'Asian, Native Hawaiian and Pacific Islander', 'Other Races or Multiple Races']
values = [27411, 42714, 25285, 13360, 13421]
total_population = 122191

percentages = [(value / total_population) * 100 for value in values]

plt.figure(figsize=(10, 5))
bars = plt.bar(categories, values, color=['blue', 'green', 'red', 'yellow', 'purple'])

plt.title('Race Demographics of Dorchester Residents')
plt.xlabel('Race')
plt.ylabel('Population')
plt.xticks(rotation=45)

for bar, percentage in zip(bars, percentages):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{percentage:.2f}%', ha='center', va='bottom', color='black')

plt.show()
```
