```python
import matplotlib.pyplot as plt

# Data for Race Demographics in Downtown
categories_race = ['White', 'Black or African American', 'Hispanic or Latino',
                   'Asian, Native Hawaiian and Pacific Islander', 'Other Races or Multiple Races']
values_race = [8730, 569, 967, 3018, 484]
total_population_race = 13768

percentages_race = [(value / total_population_race) * 100 for value in values_race]

plt.figure(figsize=(10, 6))
bars_race = plt.bar(categories_race, percentages_race, color=['blue', 'green', 'red', 'yellow', 'purple'])
plt.xlabel('Race')
plt.ylabel('Percentage')
plt.title('Race Demographics of Downtown Residents (%)')
for bar in bars_race:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom')

plt.tight_layout()
plt.show()


# Data for Age Demographics in Downtown
categories_age = ['18+', '0-17']
values_age = [12975, 793]
total_population_age = sum(values_age)

percentages_age = [(value / total_population_age) * 100 for value in values_age]

plt.figure(figsize=(10, 6))
bars_age = plt.bar(categories_age, percentages_age, color=['lightblue', 'lightgreen'])
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.title('Age Distribution of Downtown Residents (%)')
for bar in bars_age:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
