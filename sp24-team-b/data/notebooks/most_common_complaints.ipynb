```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Change the file path for the csv file as needed
df = pd.read_csv('./311_2019_to_present.csv')
# Get the number of times that each complaint happens
data_values = df['reason'].value_counts()
# Get the top 5 reason counts
top_5_reason_counts = data_values.head(5)
# Get the actual labels
top_5_reasons = top_5_reason_counts.index
colors = ['blue', 'red', 'green', 'orange', 'yellow']
fig = plt.figure(figsize= (18, 10))
plt.bar(top_5_reasons, top_5_reason_counts, width=0.4, color=colors)
plt.tight_layout
plt.title("Top 5 Reasons Complaints Were Made by Residents")
plt.ylabel("Frequency")
plt.xlabel("Reasons")
```

```python
# Get the number of times that each complaint happens
data_values = df['type'].value_counts()
# Get the top 5 reason counts
top_5_reason_counts = data_values.head(5)
# Get the actual labels
top_5_reasons = top_5_reason_counts.index
#Just adds some color to the graphs
colors = ['blue', 'red', 'green', 'orange', 'yellow']
fig = plt.figure(figsize= (18, 10))
plt.bar(top_5_reasons, top_5_reason_counts, color=colors, width=0.4)
plt.tight_layout
plt.title("Top 5 Complaint Types Made by Residents")
plt.ylabel("Number of Complaints")
plt.xlabel("Types of Complaints")
```
