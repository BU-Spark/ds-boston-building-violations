```python
import matplotlib.pyplot as plt
```

```python
# Downtown: #1 Neighborhoods with Most Building Violations (Normalized)
# Demographic: Race
categories = ['White', 'Black or African American', 'Hispanic or Latino',
              'Asian, Native Hawaiian and Pacific Islander', 'Other Races or Multiple Races']
values = [8730, 569, 967, 3018, 484]
total_population = 13768

percentages = [(value / total_population) * 100 for value in values]

plt.figure(figsize=(10, 7))
wedges, texts, autotexts = plt.pie(values, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'yellow', 'purple'], startangle=140)
plt.legend(wedges, categories, title="Races", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('Race Demographics of Downtown Residents')
plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.show()

# Demographic: Age
categories = ['18+', '0-17']
values = [12975, 793]
total_population = sum(values)

percentages = [(value / total_population) * 100 for value in values]

plt.figure(figsize=(10, 7))
wedges, texts, autotexts = plt.pie(values, autopct='%1.1f%%', colors=['lightblue', 'lightgreen'], startangle=90)
plt.legend(wedges, categories, title="Age Groups", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('Age Distribution of Downtown Residents')
plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.show()
```
