```python
import pandas as pd
import matplotlib.pyplot as plt

def count_properties_by_year(file_path):
    # Load the data
    data = pd.read_csv(file_path)
    
    # Filter the columns if there are many (assuming column names are 'CITY' and 'YR_BUILT')
    data = data[['CITY', 'YR_BUILT']]
    
    # Drop rows where Year Built is not specified if any
    data = data.dropna(subset=['YR_BUILT'])
    
    # Convert 'Year Built' to integer if not already
    data['YR_BUILT'] = data['YR_BUILT'].astype(int)
    
    # Group by 'City' and 'Year Built' and count occurrences
    result = data.groupby(['CITY', 'YR_BUILT']).size().reset_index(name='Count')
    
    return result

def plot_data(result):
    # Filter data to include only years up to 2024
    result = result[result['YR_BUILT'] <= 2024]

    # Setting up the figure and axes for the plot
    fig, ax = plt.subplots(figsize=(15, 10))
    
    # Loop through each city to plot their data separately
    for city in result['CITY'].unique():
        city_data = result[result['CITY'] == city]
        ax.bar(city_data['YR_BUILT'], city_data['Count'], label=city)
    
    # Adding labels and title
    ax.set_xlabel('Year Built')
    ax.set_ylabel('Number of Properties')
    ax.set_title('Number of Properties Built Each Year by City (1800 to 2024)')
    
    # Set the x-axis limits from 1800 to 2024
    ax.set_xlim(1800, 2024)

    # Rotate x-axis labels for better visibility
    plt.xticks(rotation=45)
    
    # Add a legend
    plt.legend(title='City')
    
    # Adjust layout to not cut off labels
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Example usage:
file_path = '/Users/chloe/Desktop/fy2024-property-assessment-data_1_5_2024.csv'
result = count_properties_by_year(file_path)
plot_data(result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def count_properties_by_year(file_path):
    # Load the data
    data = pd.read_csv(file_path)
    
    # Filter the columns if there are many (assuming column names are 'CITY' and 'YR_BUILT')
    data = data[['CITY', 'YR_BUILT']]
    
    # Drop rows where Year Built is not specified if any
    data = data.dropna(subset=['YR_BUILT'])
    
    # Convert 'Year Built' to integer if not already
    data['YR_BUILT'] = data['YR_BUILT'].astype(int)
    
    # Group by 'City' and 'Year Built' and count occurrences
    result = data.groupby(['CITY', 'YR_BUILT']).size().reset_index(name='Count')
    
    return result

def plot_data(result):
    # Filter data to include only years up to 2024 and from 1800
    result = result[(result['YR_BUILT'] <= 2024) & (result['YR_BUILT'] >= 1800)]

    # Setting up the figure and axes for the plot
    fig, ax = plt.subplots(figsize=(15, 10))
    
    # Collect data by years and cities for plotting
    years = result['YR_BUILT'].sort_values().unique()
    cities = result['CITY'].unique()
    
    # Plotting each city data separately
    for city in cities:
        city_data = result[result['CITY'] == city]
        # Create an ordered DataFrame for plotting
        city_data = city_data.set_index('YR_BUILT').reindex(years).fillna(0)
        ax.bar(city_data.index, city_data['Count'], label=city, width=0.8)

    # Adding labels and title
    ax.set_xlabel('Year Built')
    ax.set_ylabel('Number of Properties')
    ax.set_title('Number of Properties Built Each Year by City (1800 to 2024)')

    # Set x-axis to display all years within the specified range
    ax.set_xticks(years)
    ax.set_xticklabels(years, rotation=90)  # Rotate for better readability

    # Add a legend
    plt.legend(title='City')
    
    # Adjust layout to not cut off labels
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Example usage:
file_path = '/Users/chloe/Desktop/fy2024-property-assessment-data_1_5_2024.csv'
result = count_properties_by_year(file_path)
plot_data(result)
```
