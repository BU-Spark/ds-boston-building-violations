```python
#create graph for overall cond and city

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/Users/chloe/Desktop/fy2024-property-assessment-data_1_5_2024.csv'
data = pd.read_csv(file_path)

# Count occurrences of each condition in the 'OVERALL_COND' column
overall_cond_counts = data['OVERALL_COND'].value_counts()

# Plotting the bar graph for 'OVERALL_COND'
plt.figure(figsize=(10, 6))
overall_cond_counts.plot(kind='bar')
plt.title('Distribution of Overall Conditions')
plt.xlabel('Overall Condition')
plt.ylabel('Number of Properties')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
file_path = '/Users/chloe/Desktop/fy2024-property-assessment-data_1_5_2024.csv'  # Make sure to replace this with your actual file path
data = pd.read_csv(file_path)

# Group the data by 'CITY' and 'OVERALL_COND', and count the occurrences
grouped_data = data.groupby(['CITY', 'OVERALL_COND']).size().unstack(fill_value=0)

# Select the top N cities for clarity in the graph
top_cities = data['CITY'].value_counts().head(10).index
filtered_data = grouped_data.loc[top_cities]

# Setting a more distinct color palette
colors = sns.color_palette('tab10', n_colors=len(filtered_data.columns))

# Plotting the grouped bar chart with the new color palette
plt.figure(figsize=(14, 8))
filtered_data.plot(kind='bar', width=0.8, color=colors)
plt.title('Number of Locations by Overall Condition in Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Number of Locations')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Overall Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
file_path = '/Users/chloe/Desktop/fy2024-property-assessment-data_1_5_2024.csv'  # Replace this with your actual file path
data = pd.read_csv(file_path)

# Filter the dataset for cities with up to 2000 locations
city_counts = data['CITY'].value_counts()
cities_up_to_2000 = city_counts[city_counts <= 2000].index
filtered_data_for_scale = data[data['CITY'].isin(cities_up_to_2000)]

# Group the filtered data by 'CITY' and 'OVERALL_COND', and count the occurrences
grouped_filtered_data = filtered_data_for_scale.groupby(['CITY', 'OVERALL_COND']).size().unstack(fill_value=0)

# Setting a more distinct color palette
colors = sns.color_palette('tab10', n_colors=len(grouped_filtered_data.columns))

# Plotting the grouped bar chart with the distinct color palette for the filtered data
plt.figure(figsize=(14, 8))
grouped_filtered_data.plot(kind='bar', width=0.8, color=colors, legend=False)
plt.title('Number of Locations by Overall Condition (Cities up to 2000 Locations)')
plt.xlabel('City')
plt.ylabel('Number of Locations')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Overall Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np  # Importing numpy for the square root function

# Load your dataset
file_path = '/Users/chloe/Desktop/fy2024-property-assessment-data_1_5_2024.csv'
data = pd.read_csv(file_path)

# Group the data by 'CITY' and 'OVERALL_COND', and count the occurrences
grouped_data = data.groupby(['CITY', 'OVERALL_COND']).size().unstack(fill_value=0)

# Apply square root transformation to scale down the values
#grouped_data = grouped_data.applymap(np.sqrt)

# Apply fourth root transformation to scale down the values
grouped_data = grouped_data.applymap(lambda x: x**(1/4))


# Select the top N cities for clarity in the graph
top_cities = data['CITY'].value_counts().head(10).index
filtered_data = grouped_data.loc[top_cities]

# Setting a more distinct color palette
colors = sns.color_palette('tab10', n_colors=len(filtered_data.columns))

# Plotting the grouped bar chart with the new color palette and scaled values
plt.figure(figsize=(14, 8))
filtered_data.plot(kind='bar', width=0.8, color=colors)
plt.title('Number of Locations by Overall Condition in Top 10 Cities (Scaled)')
plt.xlabel('City')
plt.ylabel('Forth Root of Number of Locations')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Overall Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
